# Feature Extraction

```python
import re

def extract_between(text, start_emoji, end_pattern=r"\n"):
    """
    Ambil teks setelah emoji tertentu hingga sebelum newline
    """
    if not text:
        return None
    pattern = rf"{re.escape(start_emoji)}\s*(.*?){end_pattern}"
    match = re.search(pattern, text, flags=re.DOTALL)
    return match.group(1).strip() if match else None


def extract_multiline_after(text, start_emoji):
    if not text:
        return None

    lines = text.split("\n")
    addresses = []
    current_address = []

    for line in lines:
        stripped = line.strip()

        if stripped.startswith(start_emoji):
            if current_address:
                addresses.append(" ".join(current_address).strip())
                current_address = []

            current_address.append(
                stripped.replace(start_emoji, "").strip()
            )
            continue

        if current_address:
            if re.match(r"[üñ•Ô∏èüìû‚òéÔ∏è‚è∞üóìÔ∏è‚úîÔ∏è#]", stripped):
                break
            current_address.append(stripped)

    if current_address:
        addresses.append(" ".join(current_address).strip())

    return "; ".join(addresses) if addresses else None



def extract_resto_name(text, max_words=9):
    if not text:
        return None

    lines = [l.strip() for l in text.split("\n")]

    for i, line in enumerate(lines):
        if line.startswith("üñ•Ô∏è"):
            for prev in reversed(lines[:i]):
                if not prev:
                    continue
                if re.search(r"reserve\s*now|order\s*now|gass", prev, re.IGNORECASE):
                    continue
                if re.match(r"[üìçüìû‚òéÔ∏è‚è∞üóìÔ∏è‚úîÔ∏è#]", prev):
                    continue

                clean = re.sub(r"[^\w\s]", "", prev)
                if 1 <= len(clean.split()) <= max_words:
                    return prev

    for i, line in enumerate(lines):
        if line.startswith("üìç"):
            for prev in reversed(lines[:i]):
                if not prev:
                    continue
                if re.match(r"[üìçüìû‚òéÔ∏è‚è∞üóìÔ∏è‚úîÔ∏è#]", prev):
                    continue

                clean = re.sub(r"[^\w\s]", "", prev)
                if 1 <= len(clean.split()) <= max_words:
                    return prev

    return None



data["nama_resto"] = data["caption"].apply(extract_resto_name)

data["akun_resto"] = data["caption"].apply(
    lambda x: extract_between(x, "üñ•Ô∏è")
)

data["alamat"] = data["caption"].apply(
    lambda x: extract_multiline_after(x, "üìç")
)

data["nomor_telepon"] = data["caption"].apply(
    lambda x: extract_between(x, "üìû")
)

data["jam_buka"] = data["caption"].apply(
    lambda x: extract_between(x, "‚è∞")
)
```

```python
def clean_caption_after_last_dot_enter(text):
    if not text:
        return text

    marker = ".\n"
    last_index = text.rfind(marker)

    if last_index == -1:
        return text.strip()

    return text[:last_index].strip()

data["caption_bersih"] = data["caption"].apply(clean_caption_after_last_dot_enter)
```

### Hapus Emoji

```python
import re
import pandas as pd

EMOJI_REGEX = re.compile(
    "["
    "\U0001F300-\U0001FAFF"  
    "\U00002700-\U000027BF"  
    "\U00002600-\U000026FF"  
    "\U00002000-\U00002BFF"  
    "]+",
    flags=re.UNICODE
)

def clean_caption(text):
    if not isinstance(text, str):
        return text

    text = text.replace("\uFE0F", "")

    text = EMOJI_REGEX.sub("", text)

    return text


data["caption_bersih"] = data["caption_bersih"].apply(clean_caption)
```

### Konversi jam_buka

```python
import re

def convert_jam_buka(text):
    if not text:
        return text

    text = text.strip()

    pattern = r"(\d{1,2})\s*(AM|PM)\s*-\s*(\d{1,2})\s*(AM|PM)"
    match = re.match(pattern, text, flags=re.IGNORECASE)

    if not match:
        return text  

    start_hour, start_ampm, end_hour, end_ampm = match.groups()

    def ampm_to_id(hour, ampm):
        ampm = ampm.upper()
        return f"{hour} pagi" if ampm == "AM" else f"{hour} malam"

    start = ampm_to_id(start_hour, start_ampm)
    end = ampm_to_id(end_hour, end_ampm)

    return f"{start} hingga {end}"

data["jam_buka"] = data["jam_buka"].apply(convert_jam_buka)
```

### Hapus CTA

```python
import re

CTA_KEYWORDS = [
    "order", "reserve", "book", "cek", "follow",
    "datang", "mampir", "buruan", "cus", "gass", "gas", "cuss", "klik", "recommended", "try", "coba"
]

CTA_PATTERN = re.compile(
    r"\b(" + "|".join(map(re.escape, CTA_KEYWORDS)) + r")\b",
    flags=re.IGNORECASE
)

def contains_cta(text):
    if not text:
        return False
    return bool(CTA_PATTERN.search(text))

data["has_cta"] = data["caption_bersih"].apply(contains_cta)
```

```python
def remove_cta_heuristic(text, max_words=20):
    if not text:
        return text

    sentences = re.split(r'(?<=[.!?])\s+', text.strip())
    cleaned_sentences = []

    for s in sentences:
        if contains_cta(s) and len(s.split()) <= max_words:
            continue
        cleaned_sentences.append(s)

    return " ".join(cleaned_sentences).strip()


data["caption_no_cta"] = data["caption_bersih"].apply(remove_cta_heuristic)
```

```python
for col in ['ig_full_name', 'ig_bio', 'ig_username']:
    if col in data.columns:
        data[col] = data[col].fillna("").astype(str).str.lower()

def has_exact_keywords(text, keywords):
    """
    Cek apakah salah satu keyword muncul sebagai kata / frasa utuh
    """
    for kw in keywords:
        pattern = r'\b' + re.escape(kw) + r'\b'
        if re.search(pattern, text):
            return True
    return False

def infer_content_type(row):

    text = " ".join([
        row.get("ig_full_name", ""),
        row.get("caption_no_cta", ""),
        row.get("ig_bio", ""),
        row.get("ig_username", "")
    ]).lower()

    categories = set()

    # =============================
    # EVENT / MARKET / BAZAAR
    # =============================
    if has_exact_keywords(text, [
        "market", "bazaar", "food market", "food bazaar",
        "christmas market", "pop up", "pop-up",
        "festival", "event kuliner", "tenant","christmas",
        "celebrating","flavours from peru"
    ]):
        categories.add("Event")

    # =============================
    # CAFE / COFFEE
    # =============================
    if has_exact_keywords(text, [
        "cafe", "coffee", "kopi", "matcha",
        "espresso", "latte",'ngopi'
    ]):
        categories.add("Cafe / Coffee Shop")

    # =============================
    # DESSERT / BAKERY
    # =============================
    if has_exact_keywords(text, [
        "dessert", "cake", "pastry", "bakery",
        "mochi", "pudding", "jelly",
        "sticky rice", "mango sagoo", "bolu","mango","fruit",
        "ice cream"
    ]):
        categories.add("Dessert / Bakery")

    # =============================
    # MINUMAN
    # =============================
    if has_exact_keywords(text, [
        "drink", "minuman", "tea", "teh",
        "teapot", "boba", "milk tea",
        "matcha drink"
    ]):
        categories.add("Minuman / Beverage")

    # =============================
    # RESTAURANT
    # =============================
    if has_exact_keywords(text, [
        "restaurant", "resto", "dining",
        "eatery", "kitchen","hotpot"
    ]):
        categories.add("Restaurant")

    # =============================
    # WESTERN / FAST FOOD
    # =============================
    if has_exact_keywords(text, [
        "burger", "beef patty",
        "smash burger", "american-style",
        "grill", "bbq", "pork","steak","pizza",
        "american"
    ]):
        categories.add("Western / Fast Food")

    # =============================
    # ASIAN
    # =============================
    if has_exact_keywords(text, [
        "asian", "sushi", "ramen",
        "thai food", "vietnamese",
        "dimsum", "japanese",
        "singapore", "singapura","japan","mandhi",
        "thai","korea","korean","nasi lemak"
    ]):
        categories.add("Asian")

    # =============================
    # STREET FOOD
    # =============================
    if has_exact_keywords(text, [
        "street food", "jajanan"
    ]):
        categories.add("Street Food")

    # =============================
    # HOTEL / STAYCATION
    # =============================
    if has_exact_keywords(text, [
        "hotel", "staycation",
        "resort", "villa",
        "citadines", "asr property"
    ]):
        categories.add("Hotel / Staycation")

    # =============================
    # OLEH-OLEH
    # =============================
    if has_exact_keywords(text, [
        "oleh-oleh", "souvenir", "gift shop"
    ]):
        categories.add("Toko Oleh-Oleh / Souvenir")

    # =============================
    # BUFFET
    # =============================
    if has_exact_keywords(text, [
        "buffet", "all you can eat"
    ]):
        categories.add("Buffet / All You Can Eat")

    # =============================
    # BAR (AMAN DARI barendo, barber, dll)
    # =============================
    if has_exact_keywords(text, [
        "bar", "lounge"
    ]):
        categories.add("Bar")

    # =============================
    # INDONESIAN FOOD
    # =============================
    if has_exact_keywords(text, [
        "bakso", "siomay", "kwetiau",
        "nusantara", "sunda", "bali",
        "kampung", "bakmi", "cimol",
        "klaten","bakwan","pastel",
        "ikan goreng dapoer nelayan",
        "nasi ulam","gehu "
    ]):
        categories.add("Indonesian Food")

    # =============================
    # KLINIK
    # =============================
    if has_exact_keywords(text, [
        "eye-care", "skin care",
        "face treatment", "clinic", "face"
    ]):
        categories.add("Klinik")

    # =============================
    # CATERING
    # =============================
    if has_exact_keywords(text, [
        "catering"
    ]):
        categories.add("Catering")

    # =============================
    # HAMPERS
    # =============================
    if has_exact_keywords(text, [
        "hampers"
    ]):
        categories.add("Hampers")

    # =============================
    # FALLBACK
    # =============================
    if not categories:
        categories.add("Unknown")

    return sorted(categories)

data["jenis_tempat"] = data.apply(infer_content_type, axis=1)
```

```python
import pandas as pd
import numpy as np

def merge_alamat_lokasi(row):
    alamat = row.get("alamat_resolved")
    lokasi = row.get("lokasi")

    if not isinstance(alamat, str) or not isinstance(lokasi, str):
        return alamat

    alamat_lower = alamat.lower()
    lokasi_lower = lokasi.lower()

    # ambil kata pertama dari lokasi
    kata_pertama_lokasi = lokasi_lower.split()[0]

    # cek apakah kata pertama lokasi sudah ada di alamat
    if kata_pertama_lokasi in alamat_lower:
        return alamat
    else:
        return f"{alamat}, {lokasi}"

df["alamat_final"] = df.apply(merge_alamat_lokasi, axis=1)
```

```python
import re
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.decomposition import LatentDirichletAllocation
import nltk
from nltk.corpus import stopwords
from Sastrawi.StopWordRemover.StopWordRemoverFactory import StopWordRemoverFactory

nltk.download("stopwords")

# stopword English
stopwords_en = stopwords.words("english")

# stopword Indonesia (Sastrawi)
factory = StopWordRemoverFactory()
stopwords_id = factory.get_stop_words()

# gabungkan
stopwords_all = list(set(stopwords_en + stopwords_id))

# stopword tambahan dari domain food vlogger
custom_stopwords = {
    'cuma','nya','menu','udah','idr','new','year','tempat','pilihan',
    'punya','mid','promo','bandung','jakarta','bali','buat','chef','person',
    'dine','pm','banget','cobain','baru','lantai','rasa','bareng','course','taste','live','pax','perfect','celebration','celebrate',
    'experience','buy','get','makanan','minuman','jadi','aja','kalo','super','mulai','buka','lengkap','flavor','stay','every','feel',
    'season','selalu','special','area','jam','makan','pake','ga','dish','highlight','many','hari','open','grand','night','available',
    'food','rekomendasi','signature','fry','banyak','goreng','campur','set','pesen','lebih','free','bakar','one','room','raman','collaboration',
    'flavour','la','dec','sore','slice','terbaru','cocok','culinary','kalian','sei','sama','langsung','sekarang','poppin','sunday','authentic','mani',
    'roast','grill','pik','dapet','varian','enak','aneka','kuah','tenant','bener','orang','beneran','bikin','makin','berbagai','ceremonial','good','series',
    'top','day','like','fav','tanggal','vibe','tempatnya','suka','hadir','autumn','bold','toast','ala'
}

# pastikan jadi SET
stopwords_all = set(stopwords_all)

# baru digabung
stopwords_all = stopwords_all.union(custom_stopwords)
```

```python
import re
import spacy
from Sastrawi.Stemmer.StemmerFactory import StemmerFactory

# spaCy English
nlp_en = spacy.load("en_core_web_sm", disable=["ner", "parser"])

# Sastrawi (Indonesia)
factory = StemmerFactory()
stemmer_id = factory.create_stemmer()

def lemmatize_caption(text):
    if not isinstance(text, str):
        return text

    text = text.lower()
    text = re.sub(r"http\S+|www\S+", "", text)   # hapus URL
    text = re.sub(r"[^a-zA-Z\s]", " ", text)     # hapus simbol
    text = re.sub(r"\s+", " ", text).strip()

    tokens = text.split()

    hasil = []

    for token in tokens:
        # jika kata Inggris ‚Üí spaCy
        if re.fullmatch(r"[a-z]+", token):
            doc = nlp_en(token)
            lemma = doc[0].lemma_
            hasil.append(lemma)
        else:
            # fallback Indonesia ‚Üí Sastrawi
            hasil.append(stemmer_id.stem(token))

    return " ".join(hasil)

data["caption_lemma"] = data["caption_no_cta"].apply(lemmatize_caption)
```

```python
# !python -m spacy download en_core_web_sm
```

```python
vectorizer = CountVectorizer(
    stop_words=list(stopwords_all),
    max_df=0.9,      # buang kata terlalu umum
    min_df=5,        # kata harus muncul minimal di 5 caption
    ngram_range=(1,2)  # unigram + bigram
)

X = vectorizer.fit_transform(data["caption_lemma"])
```

```python
n_topics = 10

lda = LatentDirichletAllocation(
    n_components=n_topics,
    random_state=42,
    learning_method="batch"
)

lda.fit(X)
```

```python
feature_names = vectorizer.get_feature_names_out()

def display_topics(model, feature_names, n_top_words=5):
    for topic_idx, topic in enumerate(model.components_):
        words = [feature_names[i] for i in topic.argsort()[:-n_top_words - 1:-1]]
        print(f"Topik {topic_idx}: {', '.join(words)}")

display_topics(lda, feature_names)
```

```python
topic_label_map = {
    0: "Tea & Matcha",
    1: "Hotel",
    2: "Seasonal",
    3: "Comfort Place & Casual Place",
    4: "Premium Meat",
    5: "Homey Place",
    6: "Indonesian Food",
    7: "Japanese Food",
    8: "Family Place",
    9: "Buffet & BBQ"
}

threshold = 0.05

doc_topic = lda.transform(X)

def topics_to_list(topic_probs, threshold=0.05, label_map=None):
    return [
        label_map[i]
        for i, prob in enumerate(topic_probs)
        if prob > threshold and i in label_map
    ]

data["topics"] = [
    topics_to_list(row, threshold, topic_label_map)
    for row in doc_topic
]
```
